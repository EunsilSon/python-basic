# while
특정 조건을 만족하면 코드를 반복시킴

- 구조
```
while 조건:
        실행 코드
```

````python
num = 0
while num <= 10 :
    print(num)
    num = num + 1
````

````python
a = [1, 2, 3, 4, 5]
i = 0

while i < len(a): # 리스트의 길이 만큼 반복
    print(a[i])
    i = i + 1
````

### break
while문 강제로 빠져나가기

````python
book = 10
people = 20

while people:
    print('축하합니다!')
    book = book - 1
    print('남은 책은 %d권 입니다.' %book)
    
    if book == 0 :
        print('끝났습니다. 다음 기회에!')
        break
````

### continue
조건으로 돌아가기

````python
a = 0
while a < 10 :
    a = a + 1
    if a % 2 == 0 : # 짝수
        continue
    print(a)
````

### 무한루프

````python
while 1 :
    print('Hello World!')
````

````python
while True:
    print('Hello World!')
````
